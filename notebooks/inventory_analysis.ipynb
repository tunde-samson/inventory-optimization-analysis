### Detecting Outliers in sales record using IQR

```python
Q1 = sales['sales_amount'].quantile(0.25)
Q3 = sales['sales_amount'].quantile(0.75)

IQR = Q3 - Q1 

lower_value = Q1 - 1.5 * IQR
high_value = Q3 + 1.5 * IQR
```

```python
outlier_data = sales[(sales['sales_amount'] < lower_value) | (sales['sales_amount'] > high_value)]
```

```python
outlier_data
```

```python
Sales_Clean = sales[(sales['sales_amount'] > lower_value) & (sales['sales_amount'] < high_value)]
```
